```python
# 7.4
things = ["mozzarella", "cinderella", "salmonella"] # Basic starting list
print("Section 7.4", things)

# 7.5 

# I tried things[1].capitalize() but it did not change the list.
# Instead, I used a list comprehension to capitalize Cinderella and then reconstructed the 'things' list. 
capital_person = [word.capitalize() for word in things if word == "cinderella"]
things = [things[0], capital_person[0], things[2]]
print("Section 7.5:", things)

# 7.6 

# Same process as above. Uses a list comprehension to fully capitalize the cheese. Then recreated the 'things' list.
capital_cheese = [word.replace(word, word.upper()) for word in things if word == "mozzarella"]
things = [capital_cheese[0], capital_person[0], things[2]]
print("Section 7.6:", things)

# 7.7

# pop() deletes the index at 2, which is the disease.
things.pop(2) 
print("Section 7.7:", things)

# 9.1
def good(*args):
    """
    Function named 'good' that uses *args to pass multiple words through to the 'people' list.
    """
    people = [*args]
    return people

print(good("Harry", "Ron", "Hermione")) # Activates the 'good' function and passes through the names.

# 9.2
def get_odds(num_range):
    """
    Generator function named 'get_odds' that generates only odd numbers within a specified range.
    Starts at 0 then in the while loop adds 1 to the number and checks if the number is odd by seeing if it has a remainder.
    """
    num = 0
    while num in num_range:
        if num % 2 != 0:
            yield num 
        num += 1

my_range = get_odds(range(10)) # Activates the 'get_odds' function and passes through a range from 0-9.

count = 0 # Counter
for x in my_range: # For loop iterates through the generator to print the odd numbers. 
    if count == 2: # Uses a counter to check which iteration it's on. Prints the value of the third iteration (set to 2 because we begin at 0).
        print(x)
        exit() # Without this it for some reason prints 7 and 9 as well.
    else:
        count += 1
        continue
```
